## Visualization Basics

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Matplotlib

### Bar chart

```python
# Dataset
languages = ['C', 'C++', 'Java', 'Python', 'R']
students = [23, 17, 29, 35, 12]
```

```python
# Drawing of plots
plt.bar(languages, students)

plt.title('Student Distribution Based on Language')
plt.xlabel('Programming Language')
plt.ylabel('Number of students')

plt.show()
```

```python
# Drawing of plots horinzontally
plt.barh(languages, students)

plt.title('Language Distribution Based on Student')
plt.ylabel('Programming Language')
plt.xlabel('Number of students')

plt.show()
```

```python
# Dataset
languages = ['C', 'C++', 'Java', 'Python', 'R']
students_2020 = [23, 17, 29, 35, 12]
students_2021 = [20, 15, 25, 40, 15]
```

```python
ind = np.arange(len(languages))
print(ind)
```

```python
width = 0.35

plt.bar(ind, students_2020, width=width, label='2020')
plt.bar(ind+width, students_2021, width=width, label='2021')

plt.xlabel('Programming Language\'s Popularity by Year')
plt.ylabel('Students Enrolled')
plt.legend()
plt.xticks(ind+width/2, languages)

plt.show()
```

```python
# Dataset
country = ['USA', 'Canada', 'GB', 'China', 'Russia']
bronze = np.array([38, 17, 26, 19, 15])
silver = np.array([37, 23, 18, 18, 10])
gold = np.array([46, 27, 26, 19, 17])
```

```python
#ind = np.arange(len(country))
ind = [x for x, _ in enumerate(country)]

plt.bar(ind, gold, width=0.8, label='Gold', color='gold', bottom=silver+bronze)
plt.bar(ind, silver, width=0.8, label='Silver', color='silver', bottom=bronze)
plt.bar(ind, bronze, width=0.8, label='Bronze', color='#CD853F')

plt.xticks(ind, country)
plt.ylabel('Medals')
plt.xlabel('Country')
plt.title('Olympic Standings')

plt.legend()
plt.show()
```

```python
# Dataset
languages = ['C', 'C++', 'Java', 'Python', 'R']
students = [23, 17, 29, 35, 12]
variance = [1, 2, 7, 5, 3]

ind = [x for x, _ in enumerate(languages)]
```

```python
# Drawing of plots
plt.bar(ind, students, yerr=variance)

plt.title('Student Distribution Based on Language')
plt.xlabel('Programming Language')
plt.ylabel('Number of students')
plt.xticks(ind, languages)

plt.show()
```

## Line chart

```python
#Dataset
x=[1, 2, 3, 4, 5]
y=[1, 2, 3, 4, 5]

plt.plot(x, y)
plt.show()
```

```python
# dataset: y = mx + c
m = 2
c = 1
x = np.linspace(1, 5, 10)
y = m * x + c
x, y

plt.plot(x, y)

plt.title('Line Plot')
plt.ylabel('y = m^x + c')
plt.xlabel('x')

plt.show()
```

```python
# dataset: y = mx + c
m = 2
c = 1
x = np.linspace(1, 5, 50)
y = m ** x + c
x, y

plt.plot(x, y)

plt.title('Line Plot')
plt.ylabel('y = m^x + c')
plt.xlabel('x')

plt.show()
```

```python
# dataset: y = mx + c
m = 2
c = 1
x = np.linspace(-1, 1, 50)
y1 = m * x + c
y2 = m ** x + c
x, y

plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=1, linestyle='--')

plt.title('Line Plot')
plt.ylabel('y = m^x + c')
plt.xlabel('x')

plt.show()
```
