# Visualization basics

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Bar Chart

```python
# Dataset 
languages = ['C', 'C++', 'Java', 'Python', 'R']
students = [23, 17, 29, 35, 12]
```

```python
# Drawing vertical bars
plt.bar(languages, students)
plt.title('Student distribution based on programming language')
plt.ylabel('Number of students')
plt.xlabel('Programming languages')
plt.show()
```

```python
# Drawing horizontal bars
plt.barh(languages, students)
plt.title('Student distribution based on programming language')
plt.ylabel('Programming languages')
plt.xlabel('Number of students')
plt.show()
```

```python
# Dataset 
languages = ['C', 'C++', 'Java', 'Python', 'R']
students_2020 = [23, 17, 29, 35, 12]
students_2021 = [25, 16, 28, 37, 16]
```

```python
# Index
ind = np.arange(len(languages))
width = 0.35
```

```python
plt.bar(ind, students_2020, width*0.95, label= '2020')
plt.bar(ind+width, students_2021, width*0.95, label= '2021')
plt.legend()
plt.title('Programming Language popularity per Year')
plt.ylabel('Students enrolled')
plt.xlabel('Programming languages')
plt.xticks((ind + width/2), languages)
plt.show()
```

```python
countries = ['USA', 'Canada', 'GB', 'China', 'Russia']
bronzes = np.array([38,17,26,19,15])
silvers = np.array([37,23,18,18,10])
golds = np.array([46,27,26,19,17])
```

```python
ind = [x for x, _ in enumerate (countries)]

plt.bar(ind, golds, width=0.8, label='Golds', color='gold', bottom=bronzes+silvers)
plt.bar(ind, silvers, width=0.8, label='Silvers', color='silver', bottom=bronzes)
plt.bar(ind, bronzes, width=0.8, label='Bronzes', color='#CD853F')

plt.xticks(ind, countries)
plt.ylabel('Medals')
plt.xlabel('Countries')
plt.title('Olympic standings')
plt.legend()

plt.show()
```
