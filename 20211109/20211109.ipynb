# Radar charts

```python
import matplotlib.pyplot as plt
import pandas as pd
from math import pi
```

```python
# Dataset
df = pd.DataFrame({
    'players':['kai', 'messi', 'ronaldo', 'neymar'],
    'tackling':[38, 25, 20, 15],
    'marking':[25, 15, 12, 19],
    'passing':[13, 45, 40, 42],
    'shooting':[21, 38, 42, 33],    
    'fitness':[33, 35, 45, 40]  
})
df
```

```python
categories = list(df[1:])[1:]
categories
```

```python
N = len(categories)
N
```

```python
values = df.loc[0].drop('players').values.flatten().tolist()
```

```python
values += values[:1]
```

```python
values
```

```python
angles = [n/ float(N)* 2 * pi for n in range(N)]
angles += angles[:1]
angles
```

```python
ax = plt.subplot(111, polar=True)
plt.xticks(angles[:-1], categories, color='blue', size=10)
ax.set_rlabel_position(0)
plt.yticks([10, 20, 30, 40], ["10", "20", "30", "40"], color='blue',size=9)
plt.ylim(0, 50)
ax.plot(angles, values, linewidth=1, linestyle='solid')
ax.fill(angles, values, 'g', alpha=0.1)
plt.show
```

```python
ax = plt.subplot(111, polar=True)
plt.xticks(angles[:-1], categories)
ax.set_rlabel_position(0)
plt.yticks([10, 20, 30, 40], ["10", "20", "30", "40"], color='blue',size=9)
plt.ylim(0, 50)
values = df.loc[1].drop('players').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label=str(df['players'][1]))
ax.fill(angles, values, 'g', alpha=0.1)
values = df.loc[0].drop('players').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label=str(df['players'][0]))
ax.fill(angles, values, 'g', alpha=0.1)
plt.legend(loc='upper right', bbox_to_anchor=(0.1,0.1))
plt.show
```

# pie chart

```python
labels = 'A', 'B', 'C', 'D', 'E'
values = [15, 30, 45, 25, 20]
explode = (0, 0.1, 0, 0, 0)
```

```python
ax1 = plt.subplot()
ax1.pie(values,explode=explode, labels=labels, autopct="%1.1f%%",
        shadow=True, startangle=90)
ax1.axis('equal')

plt.show()
```

```python
labels = list(df[1:])[1:]
values = df.loc[1].drop('players').values.flatten().tolist()
explode = (0, 0, 0.2, 0, 0)
```

```python
ax1 = plt.subplot()
ax1.pie(values,explode=explode, labels=labels, autopct="%1.1f%%",
        shadow=True, startangle=90)
ax1.axis('equal')

plt.show()
```

# Stacked area chart

```python
x =range(1, 6)
y1 = [1, 4, 6, 8, 9]
y2 = [2, 2, 7, 8, 2]
y3 = [1, 9, 6, 2, 4]

plt.stackplot(x, y1, y2, y3, labels=['A', 'B', 'C'])
plt.legend(loc='upper left')
plt.show()
```
