```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from math import pi
```

## Histogram

```python
np.random.seed(42)
x = np.random.normal(size = 1000)
# x. mean()
# x.std()
plt.hist(x)
plt.ylabel ('Frequency ')
plt.xlabel('Data')
```

## Density Plot

```python
import scipy.stats as st

plt.hist(x, density=True, bins=40, label='Data')

mn, mx = plt.xlim()
print(mn, mx)
kde_xs = np.linspace(mn, mx, 400)
kde = st.gaussian_kde(x)
plt.plot(kde_xs, kde.pdf(kde_xs), label= 'PDF')

plt.legend(loc='upper left')
plt.ylabel ('probability')
plt.xlabel('Data')
plt.title("Histogram and Density plot")
```

```python
# Seaborn A1 ternative
import seaborn as sns
sns.displot(x, bins= 40, kde=True)
```

## Box Plot

```python
np.random.seed(42)
# tempered data
spread = np.random.rand(50)*100
Centre = np.ones(25)*50
filer_high = np.random.rand(10)*100+100
filer_low = np.random.rand(10)*-100
data =np.concatenate((spread, Centre, filer_high, filer_low))
```

```python
data
```

```python
fig1, ax1 = plt.subplots()
ax1.set_title('basic plot')
ax1.boxplot(data)
```
